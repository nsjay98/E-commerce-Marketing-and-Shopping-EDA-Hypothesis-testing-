```python
from scipy.stats import skew

# Ploting KDE and analyze skewness
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data=df, x=col, fill=True)
    plt.title(f'KDE Plot for {col}')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.show()

    # Calculating skewness
    column_skewness = skew(df[col])


    # Determining skewness type
    if column_skewness > 0:
        skew_type = 'Right Skewed'
    elif column_skewness < 0:
        skew_type = 'Left Skewed'
    else:
        skew_type = 'Normal'

    # Finding the Median
    median = df[col].median()

    # Printing skewness value, skewness type, and median
    print(f"KDE Plot Analysis for {col}:")
    print(f"Skewness Value: {column_skewness}")
    print(f"Skewness Type: {skew_type}")
    print(f"The Median Value is {median}")
    print("\n")
```

```python
# Checking for outliers in each numerical column

# Initialize an empty DataFrame to store outlier details
df_outliers = pd.DataFrame(columns=['Column', 'Outlier Count', 'Outlier Percentage'])

# Checking for outliers in each numerical column
for col in numerical_cols:
    # Boxplot for each column
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=df, x=col)
    plt.title(f"Boxplot for {col}")
    plt.show()

    # Computing IQR for each column
    q1 = np.percentile(df[col], 25)
    q3 = np.percentile(df[col], 75)
    IQR = q3 - q1
    upper_limit = q3 + (1.5 * IQR)
    lower_limit = q1 - (1.5 * IQR)

    '''
    The values less than the lower limit and the values greater
    than the upper limit are considered outliers
    '''
    outliers_of_col = df[(df[col] < lower_limit) | (df[col] > upper_limit)]
    k = len(outliers_of_col)
    total_count = len(df)

    # Creating a DataFrame with outlier information for this column
    outlier_info = pd.DataFrame({
        'Column': [col],
        'Outlier Count': [k],
        'Outlier Percentage': [(k / total_count) * 100]
    })

    # Concatenating the current outlier info with the overall DataFrame
    df_outliers = pd.concat([df_outliers, outlier_info], ignore_index=True)

    print(f"There are {k} outliers in the '{col}' column")
    print()
    print("-------" * 10)
```

```python
# Monthly Sessions: Buyers vs Non-Buyers

# Defining the annotation function
def annotate_plot(dataframe, column, color, ax):
    for i, point in dataframe.iterrows():
        ax.text(point['Month'], point[column], str(point[column]), color='grey', ha='center', va='bottom')

# Group data by Month and Revenue, counting buyers and non-buyers
month_customer = df[df['Revenue'] == "Buyers"].groupby('Month')['Revenue'].agg(Buyers='count').reset_index()
month_customer['Non-Buyers'] = df[df['Revenue'] == "Not Buyers"].groupby('Month')['Revenue'].agg('count').values

# Initialize the plot
plt.figure(figsize=(12, 8))
ax = plt.gca()  # Get current axis

# Plot and annotate for Buyers
buyers_color = sns.color_palette('pastel')[2]
sns.lineplot(data=month_customer, x='Month', y='Buyers', label='Buyers', color=buyers_color, marker='o', ax=ax)
annotate_plot(month_customer, 'Buyers', buyers_color, ax)

# Plot and annotate for Non-Buyers
non_buyers_color = sns.color_palette('pastel')[3]
sns.lineplot(data=month_customer, x='Month', y='Non-Buyers', label='Non-Buyers', color=non_buyers_color, marker='o', ax=ax)
annotate_plot(month_customer, 'Non-Buyers', non_buyers_color, ax)

# Set labels, title, and legend
plt.ylabel('Session Count')
plt.title('Monthly Sessions: Buyers vs Non-Buyers')
plt.legend(title="Session Type")


plt.show()
```

```python
# Summary of  Each Page Category

page_categories = ["Administrative", "ProductRelated", "Informational"]
engagement_metrics = ["Administrative_Duration", "Informational_Duration", "ProductRelated_Duration", "ExitRates", "BounceRates"]

segment_data = df[page_categories + engagement_metrics]

for category in page_categories:

    category_data = segment_data[segment_data[category] > 0]

    # Summarize page views
    page_views = category_data[category].sum()

    # Summarize page durations
    avg_duration = category_data[f"{category}_Duration"].mean()

    # Summarize exit rates
    avg_exit_rate = category_data["ExitRates"].mean()

    # Summarize bounce rates
    avg_bounce_rate = category_data["BounceRates"].mean()

    print(f"\nSummary for Page Category: {category}")
    print(f"- Total Page Views: {page_views}")
    print(f"- Average Page Duration: {avg_duration:.2f} ")
    print(f"- Average Exit Rate: {avg_exit_rate:.2f}")
    print(f"- Average Bounce Rate: {avg_bounce_rate:.2f}")
```

```python
#  Average Page Views by Revenue

# Creating a new DataFrame to hold the average page views
average_page_views = pd.DataFrame()

# Group by Revenue and calculate the average page views for each page category
for category in page_categories:
    avg_page_views = df.groupby('Revenue')[category].mean().reset_index()
    avg_page_views.rename(columns={category: f"Avg_{category}_Views"}, inplace=True)

    # Merge with the main DataFrame on Revenue
    if average_page_views.empty:
        average_page_views = avg_page_views
    else:
        average_page_views = pd.merge(average_page_views, avg_page_views, on='Revenue')


print("Average Page Views by Revenue:")
average_page_views
```

Observations:
* Administrative Views: Buyers have an average of 3.39 views, compared to 2.14 for non-buyers.
* Product-Related Views: Buyers average 48.21 views, while non-buyers average 29.05 views.
* Informational Views: Buyers average 0.79 views, whereas non-buyers average 0.46 views.

```python
# Average Engagement Metrics by Revenue
revenue_engagement = df.groupby('Revenue').agg(
    Avg_Admin_Duration=('Administrative_Duration', 'mean'),
    Avg_Info_Duration=('Informational_Duration', 'mean'),
    Avg_Prod_Duration=('ProductRelated_Duration', 'mean'),
).reset_index()

# Converting durations from seconds to minutes
revenue_engagement['Avg_Admin_Duration'] /= 60
revenue_engagement['Avg_Info_Duration'] /= 60
revenue_engagement['Avg_Prod_Duration'] /= 60

print("Average Engagement Metrics by Revenue (converted to minutes):")
revenue_engagement
```

Observations:
* Buyers spend more time on administrative pages with an average duration of 1.99 minutes compared to 1.24 minutes for non-buyers.
